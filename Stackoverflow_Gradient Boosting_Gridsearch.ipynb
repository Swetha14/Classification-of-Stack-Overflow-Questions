```python
import nltk
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
import string
import numpy as np

stopwords = nltk.corpus.stopwords.words('english')
wn = nltk.WordNetLemmatizer()

data = pd.read_csv("stackoverflow_reduced.csv")

def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [wn.lemmatize(word) for word in tokens if word not in stopwords]
    return text
```

```python
data.head()
```

```python
tfidf_vect = TfidfVectorizer(analyzer=clean_text, min_df=2, norm='l2', encoding='latin-1', ngram_range=(1, 2))
X_tfidf = tfidf_vect.fit_transform(data['post'])
X_tfidf_feat = pd.DataFrame(X_tfidf.toarray())
X_tfidf_feat
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
```

```python
gb = GradientBoostingClassifier()
param = {
    'n_estimators': [150], 
    'max_depth': [11],
    'learning_rate': [0.1]
}

clf = GridSearchCV(gb, param, cv=3, n_jobs=-1)
cv_fit = clf.fit(X_tfidf_feat, data['tags'])
results = pd.DataFrame(cv_fit.cv_results_)
```
